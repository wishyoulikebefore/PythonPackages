```python
def conn_try_again(function, wait_time=3):
    """
    重连：用作装饰器
    """
    __retries = 3
    def wrapped(*args, **kwargs):
        nonlocal __retries
        try:
            return function(*args, **kwargs)
        except:
            if __retries > 0:
                __retries -= 1
                time.sleep(wait_time)
                print("程序将在%s秒后重新运行" %(wait_time))
                return wrapped(*args, **kwargs)
            else:
                print("网络状况可能不佳")
                raise Exception
    return wrapped
```

```python
@conn_try_again
def get_data(item):
    stock_code = item["stock_code"]
    stock_name = item["stock_name"]
    df = ts.get_k_data(stock_code, "2018-10-08")
    try:
        df = df.set_index("date")
    except KeyError as e:
        print("%s %s没有数据" %(stock_code,stock_name))
        return df
    new_df = df.iloc[1:,:]
    #也可以用diff
    new_df["pre_close"] = df.iloc[:-1,:]["close"].tolist()
    new_df[stock_name]=(new_df["close"]/new_df["pre_close"]-1)*100
    new_df[stock_name] = new_df[stock_name].round(decimals=2)
    trim_df = new_df[[stock_name]]
    time.sleep(1+random.random())
    return trim_df

stock_mongo = StockConceptMongo()
stockMat = pd.DataFrame([])
count = 0
for item in stock_mongo.collection.find():
    trim_df = get_data(item)
    if not trim_df.empty:
        stockMat = trim_df.join(stockMat,how="outer")
    count += 1
    if count % 100 == 0:
        print("已完成%s个" %(count))
stockMat.to_csv("D:/hist.txt")
```
