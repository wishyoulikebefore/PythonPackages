clustermap(data, pivot_kws=None, method='average', metric='euclidean', z_score=None, standard_scale=None, figsize=None, cbar_kws=None, row_cluster=True, col_cluster=True, row_linkage=None, col_linkage=None, row_colors=None, col_colors=None, mask=None, **kwargs)

>metric：距离度量，默认欧几里得距离

>z_score：0/1

>standard_scale：0/1

>{row,col}_cluster : bool，是否对行或列聚类

>{row,col}_colors : list-like or pandas DataFrame/Series，评估群体内数据是否聚在一起

```python
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset("iris")
species = iris.pop("species")
sns.clustermap(iris)
```

```python
sns.clustermap(iris, cmap="mako", robust=True)
```

```python
sns.clustermap(iris, col_cluster=False)
```

```python
sns.clustermap(iris, standard_scale=1)
```

```python
sns.clustermap(iris, z_score=0)
```

```python
lut = dict(zip(species.unique(), "rbg"))
row_colors = species.map(lut)
sns.clustermap(iris, row_colors=row_colors)
```
