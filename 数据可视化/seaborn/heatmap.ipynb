heatmap(data, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', 
        annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, 
        square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)
>data : 要显示的数据

>vmin, vmax : 颜色列表取值范围

>cmap : 颜色列表

>center : 色彩中心值

>annot : 如果为True，讲数值写入单元格

>fmt : 表格里显示数据的类型 （“.0%”，”.3”）

>linewidths : 划分每个单元格线的宽度

>linecolor : 划分每个单元格线的颜色

```python
import seaborn as sns
%matplotlib inline

uniform_data = np.random.rand(10, 12)
sns.heatmap(uniform_data, vmin=0, vmax=1)
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
flights
```

```python
sns.heatmap(flights, cbar_kws={"orientation": "horizontal"})
```

```python
sns.heatmap(flights, annot=True, fmt="d")   
```

```python
sns.heatmap(flights, linewidths=.5)  
```

```python
sns.heatmap(flights, cmap="YlGnBu")
```

```python
 sns.heatmap(flights, center=flights.loc["January", 1955])
```
