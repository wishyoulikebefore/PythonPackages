```python
from pyecharts import Bar,Bar3D
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crashes = sns.load_dataset("car_crashes")
```

```python
crashes.head()
```

# 堆叠柱状图（pyecharts）
> is_stack=True

```python
bar = Bar("堆叠柱状图示例")
crashes = crashes.sort_values(by="speeding")
bar.add("speeding", crashes["abbrev"], crashes["speeding"], is_stack=True)
bar.add("alcohol", crashes["abbrev"], crashes["alcohol"], is_stack=True)
```

# 双层柱状图 (pyecharts)
> is_stack=True

```python
bar = Bar("堆叠柱状图示例")
crashes = crashes.sort_values(by="speeding")
bar.add("speeding", crashes["abbrev"], crashes["speeding"], is_stack=True, mark_point=["average"])
bar.add("alcohol", crashes["abbrev"], -crashes["alcohol"], is_stack=True,mark_line=["min", "max"])
```

# bar_table（mpl）

```python
data = [[ 66386, 174296,  75131, 577908,  32015],
        [ 58230, 381139,  78045,  99308, 160454],
        [ 89135,  80552, 152558, 497981, 603535],
        [ 78415,  81858, 150656, 193263,  69638],
        [139361, 331509, 343164, 781380,  52269]]
columns = ('Freeze', 'Wind', 'Flood', 'Quake', 'Hail')
rows = ['%d year' % x for x in (100, 50, 20, 10, 5)]
values = np.arange(0, 2500, 500)
value_increment = 1000
colors = plt.cm.BuPu(np.linspace(0, 0.5, len(rows)))
n_rows = len(data)
index = np.arange(len(columns)) + 0.3
bar_width = 0.4

y_offset = np.zeros(len(columns))
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row],label=rows[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset])
# Reverse colors and text labels to display the last value at the top.
colors = colors[::-1]
cell_text.reverse()

the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      rowColours=colors,
                      colLabels=columns,
                      loc='bottom')

plt.subplots_adjust(left=0.2, bottom=0.2)
plt.ylabel("Loss in ${0}'s".format(value_increment))
plt.yticks(values * value_increment, ['%d' % val for val in values])
plt.xticks([])
plt.title('Loss by Disaster')
plt.legend(loc="upper left")
plt.show()
```

# 3D barplot（mpl）

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

def PCA_analysis_3D(data,save_dir,title):
    T = data.index.str.contains("T")
    N = data.index.str.contains("N")
    pca = PCA(n_components=3)
    pca.fit(data)
    dim1_evr,dim2_evr,dim3_evr = pca.explained_variance_ratio_
    transformed_data = pca.transformed(data)
    T_data = transformed_data[T,:]
    N_data = transformed_data[N,:]
    f = plt.figure(1,figsize=(8,8))
    ax = Axes3D(f,elev=-150,azim=100)
    ax.scatter(T_data[:,0],T_data[:,1],T_data[:,2],color="red",label="Tumor")
    ax.scatter(N_data[:,0],N_data[:,1],N_data[:,2],color="blue",label="Normal")
    ax.set_xlabel("1st eigenvector: %s" %(round(dim1_evr,4)))
    ax.set_ylabel("2nd eigenvector: %s" %(round(dim2_evr,4)))
    ax.set_zlabel("3rd eigenvector: %s" %(round(dim3_evr,4)))
    ax.set_title(title)
    ax.legend()
    f.savefig("%s/%s_PCA_d3.png" %(save_dir,title),dpi=100,bbox_inches="tight")
    
```
