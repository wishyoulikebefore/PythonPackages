# HDF
分层数据文件(Hierarchical Data File)，高效存储和分发<font color="red">科学数据</font>的新型数据格式

## 特点

1. 自述性：元数据在文件内

2. 通用性：许多数据类型都可以被嵌入在一个HDF文件里

3. 灵活性：HDF允许用户把相关的数据对象组合在一起，放到一个分层结构中，向数据对象添加描述和标签

4. 扩展性：HDF极易容纳将来新增加的数据模式，容易与其他标准格式兼容。

5. 跨平台性：HDF是一个与平台无关的文件格式。HDF文件无需任何转换就可以在不同平台上使用。

# HDF5
是一种存储相同类型数值的大数组的机制

适用于可被层次性组织且数据集需要被元数据标记的数据模型

## 改进部分

## 文件组织

以分层结构组织，其中包含两个主要结构：<font color="red">组</font>和<font color="red">数据集</font>
> HDF5 group：层次性容器，可以包含数据集和子组

> HDF5 dataset：一种数组型对象，在磁盘上保存数值类型的数据

> 特征：自定义元数据信息，可被附加在数据集（以及组）上

与UNIX目录和文件一样，HDF5文件中的对象通常通过给出其绝对路径名来描述
> /表示根组

> /foo表示根组中名为foo的成员

> /foo/zoo表示组foo的成员，而组foo则是根组的成员。
